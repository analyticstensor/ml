```python
# Create single plot

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1,2,3,4,50],[1,4,9,14,25], label='food')
ax.plot([3,56,44,56,34],[3,4,5,63,3], label='accessories')
ax.plot([45,86,34,67,78],[10,12,23,33,31], label='misc')
ax.set_xlabel('Items')
ax.set_ylabel('Prices')
ax.set_title('House Hold expenses')
ax.legend()
```

```python
# Create two plot
fig,(a1, a2) = plt.subplots(1,2,sharey=True)
a1.plot([1,2,3,4,5],[10,20,30,40,50])
a1.plot([11,22,33,44,45],[12,21,31,41,51])
a1.set_title("Plot A")
a2.set_title("Plot B")
a2.scatter([2,4,6,8,10],[12,15,20,14,30])
```

```python
# Create four plot
fig, ax = plt.subplots(2,2, sharex=True, sharey=True)
ax[0,0].plot([1,2,3,4,50],[1,4,9,14,25], color='k')
ax[0,1].plot([3,56,44,56,34],[3,4,5,63,3])
ax[1,0].plot([45,86,34,67,78],[10,12,23,33,31])
ax[1,1].plot([0,45,43,26,34],[3,4,5,23,35],color='y')
ax[0,0].set_title("Food")
ax[0,1].set_title("Accessories")
ax[1,0].set_title("Misc")
ax[1,1].set_title("Stationery")


# Set common labels
fig.text(0.5, 0.04, 'Categories', ha='center', va='center')
fig.text(0.06, 0.5, 'Prices', ha='center', va='center', rotation='vertical')

# Set title
fig.suptitle('House Hold Expenses', fontsize=16)

# Save the plot as jpeg file
plt.savefig("/tmp/household_expenses.jpeg")
```
