#### 2.1 Matplotlib
Matplotlib is a desktop plotting package designed for creating publication-quality plots. Plots also know as visualization is very important in data analysis. Plots are important for:-  
* identifying anomolies or outliers
* building interactive visualization
* generating ideas for models
* identifying data transformation

Matplotlib was started by John Hunter in 2002 to enable a MATLAB-like plotting interface in Python. The matplotlib and IPython communities have collaborated to simplify interactive plotting from Jupyter notebook. It supports different GUI backends on most operating systems. We can export the visualizaitons to all of the common vector and raster graphics formats such as PDF, SVG, JPG, PNG, BMP, GIF etc. Matplotlib has several add-on toolkists for data visualization that uses matplotlib for their underlying plotting such as `seaborn`.

Matplotlib has various submodules such as:
* matplotlib
* matplotlib.afm
* matplotlib.animation
* matplotlib.colors
* matplotlib.pyplot
* matplotlib.widgets

Complete list of submodules can be [found here](https://matplotlib.org/stable/api/index.html#modules).

#### 2.3 Figures and Subplots

Plots in matplotlib reside within a **Figure** object. i.e matplotlib draws/graphs our data on Figure such as windows, Juputer widgets etc. Figures contains one or more `Axes`. Axes is an area where points are specified in terms or x-y cordinates in 2D, x-y-z cordinates in 3D etc.

Figure 2.5: Parts of Figure  
![Parts of Figure](https://matplotlib.org/stable/_images/anatomy.png)

#### 2.3.1 Parts of Figure
* **Figure**: It is the whole figure. It is the top level container for all the plot elements. Figure holds the information about `Axes`, artists, and canvas. Figure can have multiple `Axes` but must have atleast one.
* **Axes**: Axes is the region of the image. As describe earlier, figure can have many Axes but a Axes object can only be in one Figure. Axes can contain either 2, 3 or more `Axis` objects. Axis is used to define the data limits. Axes has tite, an x-label and y-label.
* **Axis**: Axis are the number-line objects. It used to set the graph range and generate the ticks (i.e. marks on the axis) and ticklabels (i.e. string label).
    * Locator: Locator object is used to determine the location of the ticks.
    * Formatter: Formatter object is used to format ticklabel strings. 
* **Artist**: All the components on the figure is an artist such as Figure, Axes, and Axis object. It includes `Text` objects, `Line2D` objects, `collections` objects, `Patch` objects etc. When a figure is captured, all the artists are drawn to the canvas. Artisst are normally attached to an Axes, so Artist cannot be shared by multiple Axes or moved to other Axes.

#### 2.4 Ways of using matplotlib
There are mainly two ways of using Matplotlib. They are:
1. Object-Oriented native matplotlib API: Using object-oriented style to explicitly create figures and axes.
2. pyplot interface: Depend on pyplot (i.e. matplotlib.pyplot) to automatically create and manage figures and axes, and use pyplot functions.

**Note**: Object oriented are mainly used for non-interactive plotting. i.e. large project using functions and scripts to make a plot. We can limit the use of pyplot interface to interactive plotting such as in Jupyter notebook so that we can control our plots easily. But we can either method for plotting.

**Important**: The input to the plotting functions should be always `numpy.array` or `numpy.ma.masked_array`. We can convert other object such as `pandas` and `numpy.matrix` to `numpy.array` easily.

#### 2.4.1 Object-Oriented Style
The code snippet below shows the object oriented style to draw a plot.

```python
fig, ax = plt.subplots() # create a figure and an axes
x = [2,4,6,8,10]   # list x
data = np.asarray(x) # Convert list x to numpy array
ax.plot(data, data, label='linear') # plot the given x and y data on the axes
ax.plot(data, data**2, label='quadratic') # plot the given x and y data on the axes
ax.plot(data, data**3, label='cubic') # plot the given x and y data on the axes
ax.set_xlabel('Label for x-axis') # Create x-label to the axes
ax.set_ylabel('Label for y-axix') # Create y-label to the axes
ax.set_title("Create simple plot") # Create title to the axes
ax.legend() # Add a legend
```

#### 2.5 Adjusting spacing in subplots

matplotlib has certain padding around the outside of the subplots and spacing between subplots. The spacing is specified relative to the height and width of the plot. If we resize the plot (changing GUI window, or programmatically) then the plot will dynamically adjust itself. We can change the spacing using `subplots_adjust` method on `Figure` objects.

`subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)`

wspace and hspace are used to control the percent of the figure width and figure height rep. to use as spacing between subplots.

```python
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),\
        bins=50, color='b',alpha=0.7)
        plt.subplots_adjust(wspace=0,hspace=0)
```

Figure 2.9: Plot with no inter-subplot spacing

In the above plot, the axis labels are overlapping. matplotlib doesn't check the labels overlap, so we need to specify explicit tick location and tick lables.

#### 2.7 Ticks, Labels, and Legends
To decorate the plot we can use:
* pyplot interface (i.e. matplotlib.pyplot)
* object-oriented native matplotlib API

The `pyplot` interface has:
* `xlim`: used to control the plot range.
* `xticks`: used to control tick locations.
* `xticklabels`: used to control tick labels.

They are called in two ways:
* called with no arguments which returns the current parameter value (e.g. plt.xlim() return the current plotting range).
* called with parameters that sets the parameter value (e.g plt.xlim([0,10]) sets the x-axis range to 0 to 10).

**Setting the title, axis labels, ticks and ticklables**

We can change the x-axis ticks and labels by `set_xticks` and `set_xticklables` resp. `set_xticks` specifies matplotlib where to place the ticks along the data range. By default, those locations will also be the labels. We can set the labels with `setxtickslabels`.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), 'k')
```

Figure 2.12: Simple plot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ticks = ax.set_xticks([0,200,500,700,900])
labels = ax.set_xticklabels(['US', 'UK', 'Germany', 'Japan', 'Russia'])
ax.plot(np.random.randn(1000).cumsum(), 'k')
```

Figure 2.13: Simple plot with xticks and xtickslabels

`rotation` option in set_xticklabels is used to set the x tick labels at 30 degree rotation. `set_xlabel` is used to name the x-axis and `set_title` is used to name subplot title.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Plot of some data')
ax.set_xlabel('Locations')
ax.plot(np.random.randn(1000).cumsum(), 'k')
```

Figure 2.14: Simple plot with title and label

We can modify the y-axis by substituting y for x in the above code. The axes class has `set` method that will allow batch setting of plot properties shown in example below.

```python
prop = { 'title': 'Plot of new data',
       'xlabel': 'Locations' }
ax.set(**prop)
```

**Adding Legends**  
Legends helps to identify the plot elements. We can create a legend by passing the `label` argument when adding each piece of the plot.

```python
from numpy.random import randn

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'r', label='apple')
ax.plot(randn(1000).cumsum(), 'b--', label='samsung')
ax.plot(randn(1000).cumsum(), 'c', label='nokia')
ax.legend(loc='best')
```

Figure 2.15: Simple plot with three lines and legend

We can call `ax.legend()` or `plt.legend()` to create a legend automatically. The `legend` method has several other choices for the location `loc` argument such as best, right, center etc. `best` will choose a location that is most out of the way. 

`_nolegend_` is used to exclude one or more elements from the legend. We can see the docstring by executing the command below.

```python
ax.legend? # Display docstring of legend
```

#### 2.8 Annotations and Drawing on a subplot

We can also draw our own plot annotations such as text, arrows, or other shapes in the standard plot. We can use `text`, `arrow`, and `annotate` functions.  
`text`: draws text at given coordinates (x,y) on the plot with optional custom style. For e.g.

```python
ax.text(x, x, 'Comment of the plot', family='monospace', fontsize=8)
```

`annotate`: we can draw labels at the indicated x and y coordinates. [Check the example](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html#examples-using-matplotlib-pyplot-annotate)  

matplotlib has objects that represent many common shapes which is referred as `patches`. Some shapes like rectangle, circle can be found in `matplotlib.pyplot` but a full set is found in `matplotlib.patches`.  

To add a shape to a plot we need to create the patch object and add it to the subplot by calling it. The code snippet shows adding a shape in the plot.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rec = plt.Rectangle((0.3, 0.8), 0.5, 0.4, color='b', alpha=0.2)
cir = plt.Circle((0.6, 0.3), 0.1, color='r', alpha=0.4)
poly = plt.Polygon([[0.2, 0.2], [0.3, 0.4], [0.1, 0.4]], color='y', alpha=0.6)

ax.add_patch(rec)
ax.add_patch(cir)
ax.add_patch(poly)
```

#### 2.9 Saving plot to file

Plot can be saved to file using `plt.savefig`. It is similar to figure object's `savefig` instance method. We can use various parameters to this method. The signature of this method is:  

`savefig(fname, dpi=None, facecolor='w', edgecolor='w',orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1, frameon=None, metadata=None)`

Table 2.4: savefig options
    
| Argument | Description |
| -------- | ----------- |
| fname | String containing a filepath or a Python file-like object. The figure format is inferred from the file extension such as .pdf for PDF, .png for PNG) |
| dpi | Resolution of figure in dots per inch; defaults to 100 out of the box but can be configured |
| facecolor, edgecolor | The color of the figure background outside the subplots e.g 'w' (white) by default) |
| format | The explicit file format to use ('png', 'pdf', 'svg' 'ps', 'eps', etc..) |
| bbox_inches | The portion of the figure to save; if 'tight' is passed it will attempt to trim the empty space around the figure |

[Check more options](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html#matplotlib-pyplot-savefig)

```python
from io import BytesIO
buffer = BytesIO()
plt.savefig(buffer)
plot_info = buffer.getvalue()
```

#### 2.10.1 Matplotlib rcparams
We can dynamically change the default `rc` settings in a python script or interactively from the python shell. rc settings are stored in a dictionary-like variable called `matplotlib.rcParams`. It is global to matplotlib package. We can also directly modify the rcParams shown below:

```python
import matplotlib as mpl

mpl.rcParams['lines.linewidth'] = 3
mpl.rcParams['lines.linestyle'] = '--'
plt.plot(x)
```

Figure 2.17: plot with different style by setting rcparams

matplotlib aslo has convenience functions for modifying `rc` settings. `matplotlib.rc` can be used to modify multiple settings by using keyword arguments as shown below:

```python
mpl.rc('lines', linewidth=4, linestyle='-.')
plt.plot(x)
```

Figure 2.18: plot with different style by modifying rc setting

We can also pass the dict into `plt.rc` method.

```python
font_options = {'family': 'monospace',
                 'weight': 'bold',
                  'size': '20'}
plt.rc('font', **font_options)
```

`matplotlib.rcdefaults` will restore the standard matplotlib default settings.  

Matplotlib uses `matplotlibrc` configuration files to customize all the properties. If a path is not specified in _style.use()_ then matplotlib will search for `matplotlibrc` in different location in order of:
1. matplotlibrc in current working directory.
2. \\$MATPLOTLIBRC if it is a file, otherwise \\$MATPLOTLIB/matplotlibrc.
3. On Linux, looks at .config/matplotlib/matplotlibrc, in other OS, looks at .matplotlib/matplotlibrc.
4. INSTALL/matplotlib/mpl-data/matplotlibrc, where INSTALL is /usr/lib/python3.7/site-packages on Linux and C:\Python37\Lib\site-packages on Windows.

After the matplotlib file is found then it won't search for other locations.

We can check the current active `matplotlibrc` file where it is loaded from by using the code below.

```python
import matplotlib

matplotlib.matplotlib_fname()
```

#### 2.10.2 Using style sheets

**Composing Styles**  
We can also create our own style and use it by calling `style.use` from the location of style sheet. We can also compose two or more style by passing a list of styles such as:  
`plt.style.use(['classic', 'seaborn-deep'])`

**Temprorary Styles**  
If we want to temprorarily use the style without changing the global styling, the style package provides a context manager for limiting the changes to a specific scope. We can use code shown below:

```python
with plt.style.context('grayscale'):
    plt.plot(np.sin(np.linspace(0,2*np.pi)), 'r-o')
plt.show()
```
